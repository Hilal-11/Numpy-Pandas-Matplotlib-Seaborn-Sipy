```python
data_points = pd.DataFrame(
    {
        '1st point': [(1 , 2) , (3 , 4) , (5 , 6) , (7 , 8)],
        '2nd point': [(9 , 10) , (11 , 12) , (13 , 14) , (15 , 16)]
    }
)
data_points
```

```python
def euclidean_distance(row):
    A = row['1st point']
    B = row['2nd point']
    return ((A[0] - B[0]) ** 2 + (A[1] - B[1]) ** 2)**0.5
data_points['distance'] = data_points.apply(euclidean_distance , axis = 1)
data_points
```
