```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
movies = pd.read_csv('imdb-top-1000.csv')
movies.sample()
```

```python
genres = movies.groupby('Genre')
# type(genre)
# genres.mean()
```

```python
# find the top genres by total rating
movies.sample()
movies.groupby('Genre').sum()['Gross'].sort_values(ascending = False).head(3)
```

```python
movies.groupby('Genre')['Gross'].sum().sort_values(ascending=False).head(5)
```

```python
# find the genre with heighest average IMDB rating
movies.groupby('Genre')['IMDB_Rating'].mean().sort_values(ascending=False).head(1)
```

```python
# find the director with the most votes/popularity
movies.head(1)
movies.groupby('Director').sum()['No_of_Votes'].sort_values(ascending=False).head(1)
```
