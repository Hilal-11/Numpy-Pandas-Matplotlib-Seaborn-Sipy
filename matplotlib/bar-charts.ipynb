```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

```

```python
# about bar charts --- 
# bar charts are used to represent categorical data. We can use bar charts to compare different categories of data. We can use the bar function to create a bar chart. The bar function takes two parameters, the first parameter is the x-axis values and the second parameter is the y-axis values. We can also specify the color of the bars and the width of the bars.
# we can also add labels to the bars by using the label parameter in the bar function.
```

```python
children = [10 , 20, 30, 40, 50]
colors = ['red', 'blue', 'green', 'orange', 'purple']

plt.bar(children, colors , color=colors, label='Children')
```

```python
# horizontal bar chart ---
# we can also create a horizontal bar chart by using the barh function. The barh

children = [10 , 20, 30, 40, 50]
colors = ['red', 'blue', 'green', 'orange', 'purple']

plt.barh(children, colors , color=colors, label='Children')
```

```python
# colors and labels
df = pd.read_csv('./batsman_season_record.csv')
df.head()
plt.bar(df['batsman'], df['2015'], width=0.2, )
plt.bar(df['batsman'], df['2016'] , width=0.2)
plt.bar(df['batsman'], df['2017'] , width=0.2)
plt.xlabel('Batsman')
plt.ylabel('Runs')
plt.title('Runs scored by batsman in 2015')
```

```python
fruits = ['Apple', 'Banana', 'Cherry', 'Date', 'Elderberry']
quantities = [10, 20, 15, 5, 8]
bar_labels = ['apple', 'banana', 'cherry', 'date', 'elderberry']
bar_colors = ['red', 'yellow', 'pink', 'brown', 'purple']
plt.bar(fruits, quantities, color=bar_colors, label=bar_labels)
plt.legend(title='Fruits')
plt.show()
```

```python
species = ('Adelie', 'Chinstrap', 'Gentoo')
sex_counts = {
    'Male': np.array([73, 34, 61]),
    'Female': np.array([73, 34, 58]),
}
width = 0.6  # the width of the bars: can also be len(x) sequence


fig, ax = plt.subplots()
bottom = np.zeros(3)

for sex, sex_count in sex_counts.items():
    p = ax.bar(species, sex_count, width, label=sex, bottom=bottom)
    bottom += sex_count

    ax.bar_label(p, label_type='center')

ax.set_title('Number of penguins by sex')
ax.legend()

plt.show()
```

```python
people = ('Tom', 'Dick', 'Harry', 'Slim', 'Jim')
y_pos = np.arange(len(people))
performance = 3 + 10 * np.random.rand(len(people))
error = np.random.rand(len(people))

plt.barh(y_pos , performance , xerr=error)
plt.yticks(y_pos, labels=people)
plt.grid()
plt.xlabel("Performance")
plt.title('How fast do you want to go today?')

hbars = plt.barh(y_pos, performance, xerr=error)
plt.bar_label(hbars)
plt.xlim(right=15)
```

```python
fruit_names = ['Coffee', 'Salted Caramel', 'Pistachio']
fruit_counts = [4000, 2000, 7000]

container = plt.bar(fruit_names, fruit_counts)
plt.xlabel("Fruit Types")
plt.ylabel("Number of Sales")
plt.title("Number of Sales by Fruit Type")
plt.bar_label(container, fmt='{:,.0f}')
```

```python
species = (
    "Adelie\n $\\mu=$3700.66g",
    "Chinstrap\n $\\mu=$3733.09g",
    "Gentoo\n $\\mu=5076.02g$",
)
weight_counts = {
    "Below": np.array([70, 31, 58]),
    "Above": np.array([82, 37, 66]),
}
width = 0.5
plt.subplots()
bottom = np.zeros(3)
for boolean , counts in weight_counts.items(): 
    container = plt.bar(species , counts , label=boolean , width=width, bottom=bottom)
    bottom += counts
    # plt.bar_label(s, label_type='center')
    plt.bar_label(container, fmt='{:,.0f}')
    
plt.bar_label(container, fmt='{:,.0f}')
plt.xlabel("Species")
plt.ylabel("Number of Penguins")
plt.title("Number of Penguins by Weight Category")
plt.legend()
plt.show()
```
