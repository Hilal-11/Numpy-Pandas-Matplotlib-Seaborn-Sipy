```python
# DATA ------ [Numerical Data and Categorical Data]
# Numerical Data --- [Continuous Data and Discrete Data] numerical data is data that can be measured and expressed in numbers. It can be further categorized into continuous data and discrete data. Continuous data can take any value within a range, while discrete data can only take specific values.

# Categorical Data --- [Ordinal Data and Nominal Data] categorical data is data that can be categorized into groups or categories. It can be further categorized into ordinal data and nominal data. Ordinal data has a natural order or ranking, while nominal data does not have a natural order and is simply used for labeling or categorization.
```

```python
# 2d line plots (x and y)

# bi-variate data Analysis--- [scatter plot and line plot]
# catagorical data analysis --- (Catagorical x-axis and numerical y-axis)
# use for time series data


price = [40000, 45000, 50000, 55000, 60000, 25000, 38000]
years = [2015, 2016, 2017, 2018, 2019, 2020, 2021]

plt.plot(years, price)
plt.xlabel('Years')
plt.ylabel('Price')
plt.title('Price of Car Over Years')
```
