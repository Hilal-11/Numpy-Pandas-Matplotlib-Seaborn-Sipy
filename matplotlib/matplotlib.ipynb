```python
# DATA ------ [Numerical Data and Categorical Data]
# Numerical Data --- [Continuous Data and Discrete Data] numerical data is data that can be measured and expressed in numbers. It can be further categorized into continuous data and discrete data. Continuous data can take any value within a range, while discrete data can only take specific values.

# Categorical Data --- [Ordinal Data and Nominal Data] categorical data is data that can be categorized into groups or categories. It can be further categorized into ordinal data and nominal data. Ordinal data has a natural order or ranking, while nominal data does not have a natural order and is simply used for labeling or categorization.
```

```python
batsman = pd.read_csv('./sharma-kohli.csv')
batsman
```

```python
plt.plot( batsman['index'], batsman['RG Sharma'])
plt.plot( batsman['index'], batsman['V Kohli'])
plt.xlabel('Runs')
plt.ylabel('Years')
plt.title('Runs vs Index')
plt.style.use('classic')
```

```python
# setting the style of the plot and colors of graph lines
# we can use any color just by wriiting the HEX code of the color or we can use the name of the color as well

# plt.rcParams['axes.prop_cycle'] = plt.cycler(color=['red', 'blue'])
plt.plot( batsman['index'], batsman['RG Sharma'], color='red')
plt.plot( batsman['index'], batsman['V Kohli'], color='blue')
plt.xlabel('Runs')
plt.ylabel('Years')
plt.title('Runs vs Index')
plt.style.use('classic')
```

```python
# style the plot lines dashed , solid and dotted etc

plt.plot( batsman['index'], batsman['RG Sharma'], linestyle='dashed', color='red')
plt.plot( batsman['index'], batsman['V Kohli'], linestyle='dotted', color='blue')
plt.xlabel('Runs')
plt.ylabel('Years')
plt.title('Runs vs Index')
plt.style.use('classic')
```

```python
# width of the plot lines can be set by using the linewidth parameter in the plot function

plt.plot( batsman['index'], batsman['RG Sharma'], color='red', linewidth=3)
plt.plot( batsman['index'], batsman['V Kohli'], color='blue', linewidth=2)
plt.xlabel('Runs')
plt.ylabel('Years')
plt.title('Runs vs Index')
plt.style.use('classic')
```

```python
# we can also use the marker to mark the data points on the graph
plt.plot( batsman['index'], batsman['RG Sharma'], color='red', linewidth=3, marker='D', markersize=10, markerfacecolor='yellow', markeredgecolor='black')
plt.plot( batsman['index'], batsman['V Kohli'], color='blue', linewidth=2, marker='D' , markersize=10 , markerfacecolor='yellow', markeredgecolor='black')
plt.xlabel('Runs')
plt.ylabel('Years')
plt.title('Runs vs Index')
plt.style.use('classic')
```

```python
# adding legends to the graph is very important as it helps us to identify which line represents which data. We can add legends to the graph by using the label parameter in the plot function and then calling the legend function.
plt.plot( batsman['index'], batsman['RG Sharma'], color='red', linewidth=3, marker='D', label='RG Sharma')
plt.plot( batsman['index'], batsman['V Kohli'], color='blue', linewidth=2, marker='D', label='V Kohli')
plt.xlabel('Runs')
plt.ylabel('Years')
plt.title('Runs vs Index')
plt.style.use('classic')
plt.legend(loc='best') # loc parameter is used to specify the location of the legend. We can use 'best' to automatically place the legend in the best location.
```

```python

# setting the limit of x and y axis is very important as it helps us to focus on the specific range of data that we want to analyze. We can set the limit of x and y axis by using the xlim and ylim functions.
price = [40000, 45000, 50000, 55000, 60000, 25000, 38000, 780000]
years = [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2025]

plt.plot(years, price)
plt.xlabel('Years')
plt.ylabel('Price')
plt.title('Price of Car Over Years')
plt.ylim(0, 80000) # to set the limit of y-axis
plt.xlim(2020, 2025) # to set the limit of x-axis
```

```python
# we can also add grid to the graph to make it more readable and easier to analyze the data. We can add grid to the graph by using the grid function.
plt.plot( batsman['index'], batsman['RG Sharma'], color='red', linewidth=3, marker='D', label='RG Sharma')
plt.plot( batsman['index'], batsman['V Kohli'], color='blue', linewidth=2, marker='D', label='V Kohli')
plt.xlabel('Runs')
plt.ylabel('Years')
plt.title('Runs vs Index')
plt.style.use('classic')
plt.legend(loc='best') # loc parameter is used to specify the location of the legend. We can use 'best' to automatically place the legend in the best location.
plt.grid() # to add grid to the graph
```
