```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

Bivariate Analysic
Numerical vs numerical
use cases finding correlations

```python
x = np.linspace(-10, 10, 50)
y = 10*x + 3 + np.random.normal(0, 300, 50)
plt.scatter(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.show()
```

```python
batter = pd.read_csv('./batter.csv')
data = batter.head(50)
plt.scatter(data['avg'], data['strike_rate'], color='red', marker='D', label='Batter')
plt.xlabel('Average')
plt.ylabel('Strike Rate')
plt.title('Average vs Strike Rate')
plt.show()
```

```python
tips = sns.load_dataset('tips')
tips
plt.scatter(tips['total_bill'], tips['tip'], label='Tips', s=tips['size']*20, alpha=0.5)
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Total Bill vs Tip')
plt.legend()
plt.show()
```
